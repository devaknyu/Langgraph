```python
from langgraph.graph import StateGraph, START, END
from langchain_google_genai import ChatGoogleGenerativeAI
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite", temperature=1.0)
```

```python
class QAstate(TypedDict):
    question: str
    answer: str
```

```python
def gen_ans(state: QAstate) -> QAstate:
    question = state['question']
    answer = model.invoke(question).content
    state['answer'] =answer

    return state
```

```python
graph = StateGraph(QAstate)
graph.add_node('gen_ans', gen_ans)
graph.add_edge(START, 'gen_ans')
graph.add_edge('gen_ans',END)

workflow = graph.compile()
```

```python
initial_state = {'question': 'what is the capital of India'}
final_state = workflow.invoke(initial_state)
print(final_state)
```
