```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
class BMIstate(TypedDict):
    height: float
    weight: float
    bmi: float
    category: str
```

```python
def bmi_calc(state: BMIstate) -> BMIstate:
    height = state['height']
    weight = state['weight']
    bmi =  weight/(height**2)
    state['bmi'] = bmi

    return state
```

```python
def bmi_category(state: BMIstate) -> BMIstate:
    bmi = state['bmi']

    if bmi < 18.5:
        state["category"] = "Underweight"
    elif 18.5 <= bmi < 25:
        state["category"] = "Normal"
    elif 25 <= bmi < 30:
        state["category"] = "Overweight"
    else:
        state["category"] = "Obese"
    
    return state
```

```python
graph = StateGraph(BMIstate)
graph.add_node('BMI_calculate', bmi_calc)
graph.add_node('bmi_category', bmi_category)
graph.add_edge(START, 'BMI_calculate')
graph.add_edge('BMI_calculate','bmi_category')
graph.add_edge('bmi_category',END)
```

```python
workflow = graph.compile()
```

```python
initial_state = {'height': 1.72, 'weight': 67}
final_state = workflow.invoke(initial_state)
print(final_state)
```
